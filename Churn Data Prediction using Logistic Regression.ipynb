### Using Logistic Regression

```python
x = customer[['tenure']] #independent variable
```

```python
y = customer[['Churn']]  #dependent variable
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split   #For model train and test
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35)  
# For 35% going to the test set observation and 65% go to the train set
```

```python
log_model.fit(x_train,y_train)
```

```python
y_pred= log_model.predict(x_test)
```

```python
y_test.head()  #actual values
```

```python
y_pred[0:5]  #predict values
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#confussion matrics
confusion_matrix(y_test,y_pred)
```
